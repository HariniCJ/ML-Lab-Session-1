```python
#1 Program to count no. of vowels and consonants in a string
def vowels_and_consonants(variable):
    count_vowels=0
    count_consonants=0

    vowels='AEIOUaeiou'

    for i in variable:
        if i in vowels:
            count_vowels+=1
        else:
            count_consonants+=1

    return count_vowels, count_consonants

#2 Program to multiply two matrices
def matrix_multiply(matrixA, matrixB):
    result = []
    for i in range(len(matrixA)):
        row = []
        for j in range(len(matrixB[0])):
            sum = 0
            for k in range(len(matrixA[0])):
                sum += matrixA[i][k] * matrixB[k][j]
            row.append(sum)
        result.append(row)
    return result
def is_multiplicable(matrixA,matrixB):
    #checks for no of coloumn in matrixA is equal to no of rows in matrixB
    return len(matrixA[0])==len(matrixB)
def print_matrix(matrix):
    for row in matrix:
        print(row)
#3 Program to find the common elements between two lists
def commonElements(list1,list2):
    common_elements=[]
    for i in list1:
        for j in list2:
            if i==j:
                common_elements.append(i)
    return common_elements
#4 Program to find transpose of matrix
def matrixInput(rows, cols):
    matrix = []
    for i in range(rows):
        row = []
        for j in range(cols):
            value = int(input(f"Enter value for row {i + 1}, column {j + 1}: "))
            row.append(value)
        matrix.append(row)
    return matrix

def transpose_matrix(matrix):
    transposed = []
    for i in range(len(matrix[0])):
        transposed_row = []
        for j in range(len(matrix)):
            transposed_row.append(matrix[j][i])
        transposed.append(transposed_row)
    return transposed

def print_matrix1(matrix):
    for row in matrix:
        print(row)

#main function where all the above functions are called
def main():
    # Question 1
    string1=input("Enter the string: ")
    vowel_count, consonant_count=vowels_and_consonants(string1)

    print("Number of vowels: ",vowel_count)
    print("Number of consonants: ",consonant_count)
    # Question 2
    matrix1=[[4,3,2],[7,8,9]]
    matrix2=[[5,6],[4,6],[7,3]]

    if is_multiplicable(matrix1,matrix2):
        totalMatrix=matrix_multiply(matrix1,matrix2)
        print("Matrix 1: ")
        print_matrix(matrix1)
        print("\nMatrix 2: ")
        print_matrix(matrix2)
        print("\nProduct of matrix: ")
        print_matrix(totalMatrix)
    else:
        print("Error")

    # Question 3
    list1=[7,8,5,4,2,1]
    list2=[2,3,5,8,9,6]
    common_elements=commonElements(list1,list2)
    print("\nThe Common elements are: ",common_elements)

    # Question 4
    rows = int(input("Enter the number of rows: "))
    cols = int(input("Enter the number of columns: "))

    print("Enter values for the matrix:")
    input_matrix = matrixInput(rows, cols)

    if input_matrix is not None:
        transposed_matrix = transpose_matrix(input_matrix)

        print("\nOriginal Matrix:")
        print_matrix1(input_matrix)

        print("\nTransposed Matrix:")
        print_matrix1(transposed_matrix)

if __name__ == "__main__":
    main()




```
